## 1.1 Matris oluşturma

```python
a = np.random.randint(0,100,(3,3))
print("transpose")
print(a.T)
print("determinant")
print(np.linalg.det(a))
print("2 katı")
b = [2]
print(a*b)
```

## 1.2 Dizi Operasyonları

```python
a = np.arange(1,50,10)
print("ortalama")
print(a.mean())
print("standart sapma")
print(a.std())
print("elemanlarının karesi")
print(a**2)
print("25 den büyük elemanlar")
print(a[a>25])
```

## 2.2 Veri İçe Aktarma ve İşleme

```python
df = sns.load_dataset("titanic")
df.head(10)
```

```python
#Eksik verileri bulun ve doldurun (örneğin, eksik yaş verilerini yaşların 
#ortalaması ile doldurun).
print("before: " + str(df["age"].isnull().sum()))
df["age"] = df["age"].fillna(value=df["age"].mean())
print("after: " + str(df["age"].isnull().sum()))
```

```python
#Cinsiyet sütunundaki kadın ve erkek yolcuların sayısını bulun.
df["sex"].value_counts()
```

```python
#Hayatta kalma oranını hesaplayın
df["survived"].value_counts()[1] * 100 / df["survived"].value_counts().sum()
```

## 2.3 Gruplama ve Görselleştirme

```python
df["pclass"].value_counts()
```

```python
df.groupby("pclass")["survived"].sum()

print("1.sınıf yolcu oranı:" + str(df["pclass"].value_counts()[1] * 100 / df["pclass"].value_counts().sum()))
print("2.sınıf yolcu oranı:" + str(df["pclass"].value_counts()[2] * 100 / df["pclass"].value_counts().sum()))
print("3.sınıf yolcu oranı:" + str(df["pclass"].value_counts()[3] * 100 / df["pclass"].value_counts().sum()))
```

```python
df.groupby("pclass")["survived"].sum()
```

```python
df.groupby("pclass")["survived"].sum()
sns.barplot(df.groupby("pclass")["survived"].sum())
```
