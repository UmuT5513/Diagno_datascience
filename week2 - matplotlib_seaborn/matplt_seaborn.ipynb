```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = [2020, 2021, 2022, 2023]
weights = [101, 110, 85, 78]
age = [35, 54, 42, 21]
name = ["ahmet", "mehmet", "murat", "sedat"]
lang = ["c++","python", "java", "go"]
votes = [34, 22, 65, 10]
```

# Matplotlib

## 1.1 

```python
plt.plot(years, weights)
```

## 1.2 ve 1.3

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))


axs[0,0].plot(years, weights)
axs[0,0].set_title("yıllara göre ağırlıkların durumu(line)")

axs[0,1].scatter(years, weights)
axs[0,1].set_title("yıllara göre ağırlıkların durumu(scatter)")

explodes = [0.0, 0.2, 0.0, 0.0]
axs[1,0].pie(votes, labels=lang, explode=explodes)
axs[1,0].set_title("Dillerin popüleritesi")

axs[1,1].bar(name, age, width=0.3, color="r", align="edge", edgecolor="green")
axs[1,1].set_title("kişilerin yaşları")

plt.savefig("saved.png", dpi=300)
```
